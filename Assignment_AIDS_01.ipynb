```python
# Mini Login Project using Nested Loops and If-Else

correct_username = "admin"
correct_password = "12345"
security_answer = "blue"

while True:

    username = input("Username enter kary: ")

    if username == correct_username:

        password = input("Password enter kary: ")

        if password == correct_password:

            print("Login successful!")

            # Security Question
            answer = input("Security Question - Your favorite color? ")

            if answer == security_answer:
                print("Access Granted")
                break
            else:
                print("Wrong Security Answer")

        else:
            print("Wrong Password")

    else:
        print("Wrong Username")

    retry = input("Dobara try karna hai? (yes/no): ")

    if retry != "yes":
        print("Program End")
        break
```
